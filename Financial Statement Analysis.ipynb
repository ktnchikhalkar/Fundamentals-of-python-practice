## Fundamental of Python <br> Financial Statement Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating the DataFrame
df = pd.DataFrame(index=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec'],
                  data={'Rev':[14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44,
                                    11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50],
                        'Exp':[12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20,
                                              3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]})
```

```python
df= df.rename_axis('Month')
```

```python
df
```

```python
#Calculating profit for every month
df['Profit']=df['Rev']-df['Exp']
df
```

```python
#Calculating profit after tax
df['ProfitAfterTax'] = df['Profit']-0.3*df['Profit']
df
```

```python
#Calculating profit margin in %
df['ProfitMargin']= np.round((df['ProfitAfterTax']/df['Rev'])*100)
df
```

```python
df['ProfitAfterTax'].mean()
```
